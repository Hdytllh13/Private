```python
import pandas as pd
import numpy as np
import random

# Set random seed agar hasil bisa direplikasi
np.random.seed(42)
random.seed(42)

# Jumlah baris sample
n = 50

# Daftar nama acak sederhana
names = [
    "Ari", "Budi", "Citra", "Dewi", "Eka", "Fajar", "Gilang", "Hana", "Indra", "Joko",
    "Kiki", "Lina", "Maya", "Nanda", "Oka", "Putra", "Rina", "Sari", "Tono", "Vina",
    "Wulan", "Yoga", "Zahra", "Bayu", "Cindy", "Dian", "Evi", "Farah", "Galih", "Hesti",
    "Ivan", "Jani", "Karin", "Luki", "Mega", "Nina", "Oni", "Pipit", "Rafi", "Sinta",
    "Tari", "Umar", "Vero", "Wati", "Yani", "Zaki", "Reno", "Sari", "Doni", "Lusi"
]

# Buat dataframe dengan variasi realistis
df_sample = pd.DataFrame({
    "employee_id": np.arange(1001, 1001 + n),
    "employee_name": random.sample(names, n),  # ambil nama unik acak
    "target_achievement": np.round(np.random.uniform(0.5, 1.5, n), 2),
    "company_tenure_years": np.round(np.random.uniform(0, 15, n), 1),
    "distance_to_office_km": np.round(np.random.uniform(1, 30, n), 1),
    "job_satisfaction": np.random.randint(1, 6, n),
    "manager_support_score": np.random.randint(1, 6, n),
    "marital_status": np.random.choice(["Married", "Single"], n, p=[0.6, 0.4]),
    "working_hours_per_week": np.random.randint(35, 65, n)
})

# Simpan ke CSV
df_sample.to_csv("employee_sample.csv", index=False)

print("✅ File berhasil dibuat: employee_sample.csv")
print(df_sample.head(10))
```
